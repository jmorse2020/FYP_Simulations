```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath = "/Users/jackmorse/Documents/University/Year 4/FYP/David-Bird/Spreadsheets_with_Comsol_results/Fig_2_data_jacketed.xlsx"
sheetname = "M=6"
data = pd.read_excel(filepath, sheet_name=sheetname)
gamma = data["gamma"]
real_x = data["real x"]
coefficients = np.polyfit(gamma, real_x, int(2))
print(f"Coefficients: {coefficients}")
```

```python
plt.plot(gamma, real_x, marker = "<", color = 'brown', label = "Data")
plt.plot(gamma, np.polyval(coefficients, gamma), color = 'red', label = "Fit")
plt.title("j value for normalised gap, jacketed")
plt.xlabel(r"Normalised gap, $\gamma$")
plt.ylabel(r"Normalised wavevector, $x_{FM}$")
plt.grid()
plt.legend()
plt.show()
```
